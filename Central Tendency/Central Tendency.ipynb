```python
import pandas as pd
import numpy as np
```

```python
path="IBM-313 Marks.xlsx"
table=pd.read_excel(path)
table
```

# Find Mean

```python
x = table['Total']
np.mean(x)
```

# Find Median

```python
np.median(x)
```

```python
import scipy
from scipy import stats
stats.mode(x)
```

# Percentile

```python
a=np.array([1,2,3,4,5])
p=np.percentile(a,50) #returns 50th percentile, eg. median
p
```

# Variance

```python
np.var(x)
```

# Population standard deviation

```python
import statistics
statistics.pstdev(x)
```

# Sample standard deviation

```python
np.std(x)
```

# Check skewness

```python
from scipy.stats import skew
skew(x)
```

```python
# since the skewness is positive, it is right skewed data
```

# Box Plot

```python
from matplotlib import pyplot as plt
plt.boxplot(x, sym='*')
plt.show()
```
